For my project, I decided that the best neighborhood in Pittsburgh would have the most trees for the freshest O2 and the most beautiful green scenery. I explored a few different metrics before coming to this conclusion, such as that perhaps the best neighborhood should have the least pollution or least amount of infrastructure, but then came to the realization that a neighborhood with the most trees would outweigh those with the sense of nature in scenery.

**The Metric:** I am measuring the sheer count of trees regardless of size or height so long as their height is greater than one (aka, not a stump). I am using a trees dataset to this do which is a record of all the trees for each street up until 2020. 

**The Best Neighborhood:** To find this, I am going through each tree in the dataset to find out which street is the maximum, having the most number of trees. In the code below, the count of trees per each street is listed. Below that, the street with the **most** trees is isolated away from the group and shown. Since the  data in the dataset is only listed as *trees per street*, the best neighborhood will be determined by that closest to the street with the most trees.

```python
import pandas as pd

file = pd.read_csv('treesdata.csv', dtype={2: str})

file = file[file['height'] > 0]

treeCount = file.groupby('street').size()

print(f"Number of trees for each neighborhood: {treeCount} \n")

maxTrees = treeCount.max()
streetMost = treeCount.idxmax()

print(f"The street with the most trees is: {streetMost} with {maxTrees} trees")
```

With Penn. Ave. identified at the street with the most trees, the best neighborhood is found below. 

- Distance to East Liberty: 372.1m
- Distance to Friendship: 418.2m
- Distance to Neighborhood Learning Alliance: 425.6m

**The Neighborhood closest is East Liberty. Therefore, East Liberty is the best neighborhood with this metric.**
