For this notebook, I decided that the best neighborhood in Pittsburgh would have the most trees for the freshest O2 and the most beautiful green scenery. I explored a few different metrics before coming to this conclusion, such as that perhaps the best neighborhood should have the least pollution or least amount of infrastructure, but then came to the realization that a neighborhood with the most trees would outweigh those with the sense of nature in scenery.

To do this I am measuring the sheer count of trees regardless of size or height so long as their height is greater than one (aka, not a stump). I am using a trees dataset to this do which is a record of all the trees for each street up until 2020. 

In order too find this, I am going through each tree in the dataset to find out which street is the maximum, having the most number of trees. In the code below, the count of trees per each street is listed. Below that, the street with the **most** trees is isolated away from the group and shown. Since the  data in the dataset is only listed as *trees per street*, the best neighborhood will be determined by that closest to the street with the most trees.

1. Read the file as pandas dataframe
2. Check if the tree is not a stump (greater than 1)
3. Group the data by street then count the quantity of each having a tree
4. Sort the count of each street from most to last (decending) then isolate the first 5
5. Print the first 5

```python
import pandas as pd

file = pd.read_csv('treesdata.csv', dtype={2: str})

file = file[file['height'] > 0]
treeCount = file.groupby('street').size()
topTrees = treeCount.sort_values(ascending = False).head(5)

print("Top streets with the most trees:")
print(topTrees)
```

- Distance to East Liberty: 372.1m
- Distance to Friendship: 418.2m
- Distance to Neighborhood Learning Alliance: 425.6m

**The Neighborhood closest is East Liberty. Therefore, East Liberty is the best neighborhood with this metric.**
