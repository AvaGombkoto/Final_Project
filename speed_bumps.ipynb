```python
import pandas as pd
import numpy
humps = pd.read_csv("speed-humps.csv",
                           index_col="hump_id",
                           parse_dates=True)
```

```python
#print the data table
humps.head(10)
```

```python
#get the different neighborhoods
neighborhoods = humps.groupby("neighborhood").count()
neighborhoods.head(10)
```

```python
#create a series or dict with the neighborhood names and the amount of bumps
#make a list of neighborhoods
the_hoods = humps.loc[:,"neighborhood"].tolist()
the_hoods = numpy.array(the_hoods)
the_hoods = numpy.unique(the_hoods)
#make a list of the number of ids
bumps = neighborhoods.loc[:,"id"].tolist()
#put the lists together in a series
speed_humps = pd.Series(bumps, index=the_hoods)
speed_humps
```

```python
#compare the neighborhoods and decide which is has the most speed bumps
most_bumps=max(bumps)
print(most_bumps)
index = bumps.index(most_bumps)
neighborhood = the_hoods[index]
print(neighborhood)
```

```python
print(neighborhood+" is the best neighborhood because it has the most speed bumps of "+str(most_bumps)+", making it more safe")
```
